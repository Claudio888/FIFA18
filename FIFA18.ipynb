```python
%matplotlib inline 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
import random 
```

```python
colors = [['#0D47A1','#1565C0','#1976D2','#1E88E5','#2196F3'],
          ['#311B92','#512DA8','#673AB7','#9575CD','#B39DDB'],
          ['#1B5E20','#388E3C','#4CAF50','#81C784','#66BB6A'],
          ['#E65100','#EF6C00','#F57C00','#FB8C00','#FF9800'],
          ['#3E2723','#4E342E','#5D4037','#6D4C41','#795548'],
          ['#BF360C','#D84315','#E64A19','#F4511E','#FF5722'],
          ['#880E4F','#AD1457','#C2185B','#D81B60','#E91E63']
         ]
```

```python
fifa = pd.read_csv('fifa18.csv',sep=',')
```

```python
fifa.head()
```

```python
fifa['Overall'].describe()
```

```python
#Limpando DF


    
colunas=['Unnamed: 0','Photo','Flag','Potential','Club Logo','Special','CAM', 'CB', 'CDM', 'CF', 'CM', 'ID',
       'LAM', 'LB', 'LCB', 'LCM', 'LDM', 'LF', 'LM', 'LS', 'LW', 'LWB',
       'Preferred Positions', 'RAM', 'RB', 'RCB', 'RCM', 'RDM', 'RF', 'RM',
       'RS', 'RW', 'RWB', 'ST']

fifa18=fifa.drop(colunas,1)

                           
fifa18.head()
```

```python
fifa18.columns

Skills=['Acceleration', 'Aggression', 'Agility', 'Balance', 'Ball control',
       'Composure', 'Crossing', 'Curve', 'Dribbling', 'Finishing',
       'Free kick accuracy', 'GK diving', 'GK handling', 'GK kicking',
       'GK positioning', 'GK reflexes', 'Heading accuracy', 'Interceptions',
       'Jumping', 'Long passing', 'Long shots', 'Marking', 'Penalties',
       'Positioning', 'Reactions', 'Short passing', 'Shot power',
       'Sliding tackle', 'Sprint speed', 'Stamina', 'Standing tackle',
       'Strength', 'Vision', 'Volleys']



fifa18.dtypes

fifa18.columns
```

```python
numelem=len(Skills)
i=0
for i in range(numelem):
    fifa18[(Skills[i])] = fifa18[(Skills[i])].convert_objects(convert_numeric=True)
    i+=1
    


```

```python
fifa18.dtypes
```

```python

fig, ax = plt.subplots()


fifa18.loc[0][Skills].plot(label='CR7',color='r',figsize=[20,10])
fifa18.loc[1][Skills].plot(label='Messi',color='b')
fifa18.loc[2][Skills].plot(label='Ney',color='y')

x_axis = range(len(Skills))#tamanho do eixo X , range dos valores 
plt.legend(loc=1, prop={'size': 10}) #Tamanho da legenda
ax.set_xticklabels(Skills,rotation='vertical',fontsize=10) #atribui a legenda os nomes dos vertices 
ax.set_xticks(x_axis)#Define os valores do eixo X, fazendo todos aparecerem neste eixo.
plt.title('Skills Properties',fontsize=30) # Titulo e tamanho do titulo


plt.show()





```

```python
idade=fifa18.groupby('Age').count()
idade.drop(idade.columns[1:40],axis=1,inplace=True)
idade.columns = ['Contagem']


ax = idade.plot(kind='bar',figsize=[20,10],color=random.choice(colors),fontsize=20)
    
for p in ax.patches:
    b=p.get_bbox()
    ax.annotate("{:.0f}".format(b.y1 + b.y0), ((b.x0 + b.x1)/2 - 0.3, b.y1 + 20))
    
plt.title('Quantidade de jogadores/idade',fontsize=30) # Titulo e tamanho do titulo
plt.xlabel('Idade',fontsize=30)

```

```python
#Jogador mais rapido 

velocidade = fifa18['Sprint speed']+fifa18['Reactions'] + fifa18['Agility'] + fifa18['Acceleration']
Attvelocidade=['Sprint speed','Reactions','Agility','Acceleration']
velocidade.argmax()

neymar = fifa18.loc[velocidade.argmax()][Attvelocidade].plot(kind='bar',figsize=[10,10],color=random.choice(colors),fontsize=20)


for p in neymar.patches:
    b=p.get_bbox()
    neymar.annotate("{:.0f}".format(b.y1 + b.y0), ((b.x0 + b.x1)/2 - 0.06, b.y1 + 1))
    
plt.title('Atributos de Velocidade - Neymar ',fontsize=20) # Titulo e tamanho do titulo
plt.xticks(rotation=0)

    
plt.show()
```
